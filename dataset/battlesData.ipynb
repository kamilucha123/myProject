```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
battles = pd.read_csv('battle_dyads.csv')
```

```python
countries = battles.groupby('attacker')['isqno'].count()
countries = countries.reset_index()
countries.columns = ['Attacker', 'Count']
sortedC = countries.sort_values(by = 'Count', ascending = False)
sortedC.iloc[0]
```

```python
show = battles['attacker'].value_counts()
show = show.reset_index()
show.columns = ['country', 'count']
plt.figure(figsize = (15,5))
plt.bar(show['country'][:4],show['count'][:4])
plt.xlabel('country')
plt.ylabel('count')
```
