```python
# get the Height column
height = cols['Height']
# parse the table into feet and inches series
height = height.str.split("'")
feet = height.str.get(0)
inch = height.str.get(1)
inch = inch.str.replace('"',"")
# calculate feets and inches values
inch = inch.apply(lambda x: float(x) * 0.0254)
feet = feet.apply(lambda x: float(x) * 0.3048 )
# create single series of new values
height = (feet + inch) * 100
```

```python
# delete the old column with Height
del(cols['Height'])
```
